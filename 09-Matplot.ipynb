```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#juyper notbookta çalıştırılırken bu gereklidir
%matplotlib inline
```

```python
age_list = np.arange(20,61,5)
age_list
```

```python
weight_lis = [65,62,67,67,65,71,78,73,75]
weight_list = np.array(weight_lis)
```

```python
plt.plot(age_list, weight_list,"r") #r:red #direkt çizim işlemini başlatır
#plt.show() önceki kod çalışmıyorsa bu eklenmelidir
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age-Weight Graph")
```

```python
my_numpy1 = np.linspace(0,15,20)
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3
my_numpy2
```

```python
plt.plot(my_numpy1,my_numpy2,"g*-") #*:noktaları yıldız olarak gösterir +: artı olarak gösterir -:çizgi şeklinde gösterir --: çizgi çizgi gösterir
```

```python
plt.subplot(1,2,1) #alt grafik #1 sıra 2 kolon 1.grafiği çiziyorum
plt.plot(my_numpy1,my_numpy2,"b--")
plt.subplot(1,2,2) 
plt.plot(my_numpy2,my_numpy1,"r--")
```

```python
my_figure = plt.figure()
my_axes = my_figure.add_axes([0.1,0.1,0.5,0.5]) #boyutu 0.5,0.5 değiştirir
my_axes.plot(my_numpy1,my_numpy2,"g")
my_axes.set_xlabel("X Data")
my_axes.set_ylabel("Y Data")
my_axes.set_title("Graph Title")
```

```python
my_figure2 = plt.figure()

my_axes3 = my_figure2.add_axes([0.1,0.1,1,0.9]) #boyutu 0.5,0.5 değiştirir
my_axes3.plot(my_numpy1,my_numpy2,"r*")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Large Graph")

my_axes2 = my_figure2.add_axes([0.25,0.4,0.4,0.4])
my_axes2.plot(my_numpy1,my_numpy2,"g*")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small")
my_axes2.set_title("Small Graph")
```

```python
(my_fig,my_axe) = plt.subplots()
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 2
my_axe.plot(my_numpy3,my_numpy4, color = "#CD621D", alpha=1 )
my_axe.plot(my_numpy4,my_numpy3, color = "#B81DCD", alpha= 0.5)
```

```python
(new_fig,new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3+2, color="#B81DCD" , linewidth = 2)
new_axe.plot(my_numpy3, my_numpy3+3, color="blue" , linewidth = 1.4)
new_axe.plot(my_numpy3, my_numpy3+4, color="red" , linewidth = 1.4, linestyle = "-.")
new_axe.plot(my_numpy3, my_numpy3+5, color="gray" , linewidth = 3, linestyle = ":")
new_axe.plot(my_numpy3, my_numpy3+6, color="yellow" , linewidth = 1.4, linestyle = "--", marker = "o")
new_axe.plot(my_numpy3, my_numpy3+7, color="black" , linewidth = 1.4,linestyle = "--",  marker = "+", markersize = 8)
```
